```python
matches = pd.read_csv("IPL Matches 2008-2020.csv")
deliveries = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
#to calc. total runs
deliveries.groupby(['id', 'inning']).sum()['total_runs'].reset_index()
```

```python
total_score_ds = deliveries.groupby(['id', 'inning']).sum()['total_runs'].reset_index()
```

```python
total_score_ds = total_score_ds[total_score_ds['inning']==1]#since, we want to pred. in the 2nd innings itself.
```

```python
matches.merge(total_score_ds[['id', 'total_runs']], left_on='id', right_on='id')
```

```python
matches = matches.merge(total_score_ds[['id', 'total_runs']], left_on='id', right_on='id')
```

```python
matches['team1'].unique()#we need to remove teams who are not playing right now.
```

```python
#forming a list, which is to be there in the dataset.
teams=[
    'Royal Challengers Bangalore', 
    'Kings XI Punjab',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Rajasthan Royals', 
    'Chennai Super Kings',
    'Sunrisers Hyderabad',
    'Delhi Capitals'
]
```

```python
matches['team1']=matches['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
matches['team2']=matches['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
```

```python
matches['team1']=matches['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
matches['team2']=matches['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
matches=matches[matches['team1'].isin(teams)]
matches=matches[matches['team2'].isin(teams)]
```

```python
matches['team1'].unique()
```

```python
matches.shape
```

```python
matches
```

```python
matches = matches[['id', 'venue', 'total_runs', 'winner']]
```

```python
matches
```

```python
delivery = matches.merge(deliveries, on='id')
```

```python
delivery
```

```python
delivery['inning'].unique()
```

```python
delivery = delivery[delivery['inning']==2]
```

```python
delivery
```

```python
delivery['current_score'] = delivery.groupby('id').cumsum()['total_runs_y']
```

```python
delivery
```

```python
delivery['runs_left']=delivery['total_runs_x']-delivery['current_score']+1
```

```python
delivery
```

```python
delivery['balls_left'] = 126 - ((delivery['over']+1)*6 + delivery['ball'])
```

```python
delivery
```

```python
#to obtain wickets left
wickets = delivery.groupby('id').cumsum()['is_wicket'].values
delivery['wicket_left']=10-wickets
```

```python
delivery
```

```python
#to obtain CRR
delivery['CRR'] = delivery['current_score']*6/(120-delivery['balls_left'])
```

```python
delivery
```

```python
id_to_drop = 335982 #not complete info. for this dataset
delivery = delivery[delivery['id'] != id_to_drop]
```

```python
delivery
```

```python
#to obtain RRR
delivery['RRR'] = delivery['runs_left']*6/delivery['balls_left']
```

```python
delivery
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery['result'] = delivery.apply(result, axis=1)
```

```python
delivery.head()
```

```python
final = delivery[['batting_team', 'bowling_team', 'venue', 'runs_left', 'balls_left', 'wicket_left', 'total_runs_x', 'CRR', 'RRR', 'result']]
```

```python
final = final.sample(final.shape[0])
```

```python
final.dropna(inplace=True)
```

```python
final = final[final['balls_left']!=0]
```

```python
final
```

```python
X=final.iloc[:,:-1]
y=final.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
#since, three columns are string, so we need to apply OneHotencoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse=False, drop='first'), ['batting_team', 'bowling_team', 'venue'])], 
    remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

```python
pipe = Pipeline(steps=[('step1', trf),
               ('step2', LogisticRegression(solver='liblinear'))])
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
